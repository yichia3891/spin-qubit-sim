```python
import numpy as np
import qutip as qt
import matplotlib.pyplot as plt

Delta, Omega = 0.0, 2*np.pi*1.0
H = Delta/2 * qt.sigmaz() + Omega/2 * qt.sigmax()

psi0 = qt.basis(2, 0)
tlist = np.linspace(0, 2, 200)
result = qt.sesolve(H, psi0, tlist, e_ops=[qt.sigmaz()])

plt.plot(tlist, result.expect[0])
plt.xlabel('Time (µs)'); plt.ylabel(r'$\langle\sigma_z\rangle$')
plt.title('Rabi oscillation of a single spin qubit')
plt.show()
```
